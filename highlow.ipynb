```typescript
type LevelDetails = {
    tries: number;
    min: number;
    max: number;
};

const levels: LevelDetails[] = [
    { 'tries': 3, 'min': 1, 'max': 10 },
    { 'tries': 5, 'min': 1, 'max': 100 },
    { 'tries': 10, 'min': 1, 'max': 1000 },
];

function makeGuess() {
    // ask the user to make a guess
    let guess = Number(prompt("Guess a number: "));
    return guess;
}

function checkGuess(guess: number, answer: number) {
    if (guess < answer) {
        console.log("Too Low.");
    } else if (guess > answer) {
        console.log("Too High.");
    }
    return guess === answer;
}

function generateAnswer(level: LevelDetails) {
    // generate a random number between min and max
    let answer = Math.floor(Math.random() * level.max + level.min);
    //console.log(answer);
    return answer;
}

function displayResult(win: boolean) {
    if (win) {
        console.log("YOU WIN!");
    } else {
        console.log("You lose...");
    }
}

function guessLoop(answer, maxGuesses) {
    // can be done maxGuesses
    let numGuesses: number = 0;
    let win: boolean;
    do {
        let guess = makeGuess();
        console.log(`You guessed: ${guess}`);
        win = checkGuess(guess, answer);

        numGuesses++;
    } while (numGuesses < maxGuesses && !win);

    return win;
}

function playHighLow() {
    const maxGuesses: number = 3;
    let currentLevel: number = 1;

    let win: boolean;
    let userInput: string = 'N';
    do {
        // generated once per game
        let answer = generateAnswer(levels[currentLevel - 1]);

        win = guessLoop(answer, maxGuesses);

        displayResult(win);

        // ask user to continue on win
        if (win) {
            userInput = prompt("Do you want to continue? (Y/N) ");
            if (userInput === 'Y') {
                currentLevel++;
            }
        }
    } while(win && userInput === 'Y' && currentLevel <= levels.length);
}

// start the game
playHighLow();
```
