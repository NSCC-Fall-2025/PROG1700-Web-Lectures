```typescript
function makeGuess() {
    // ask the user to make a guess
    let guess = Number(prompt("Guess a number: "));
    return guess;
}

function checkGuess(guess: number, answer: number) {
    if (guess < answer) {
        console.log("Too Low.");
    } else if (guess > answer) {
        console.log("Too High.");
    } else {
        console.log("You Guessed It!");
    }
    return guess === answer;
}

function generateAnswer() {
    // generate a random number between 1 and 10
    let answer = Math.floor(Math.random() * 10 + 1);
    console.log(answer);
    return answer;
}

function playHighLow() {
    const maxGuesses: number = 3;

    // generated once per game
    let answer = generateAnswer();

    // can be done maxGuesses
    let numGuesses: number = 0;
    let win: boolean;
    do {
        let guess = makeGuess();
        win = checkGuess(guess, answer);

        numGuesses++;
    } while (numGuesses < maxGuesses && !win);

    if (win) {
        console.log("YOU WIN!");
    } else {
        console.log("You lose...");
    }
}

playHighLow();
```
