```typescript
// read from a text file (old way)
import * as fs from 'node:fs';

fs.readdirSync('.');
fs.existsSync('emojys.ts');
fs.copyFileSync('crappydb.csv', 'crappydb.bkp');

const content = fs.readFileSync('crappydb.csv', 'utf-8');
content.split('\r\n');

fs.writeFileSync('sample.txt', 'some stuff for the file', 'utf-8');
fs.writeFileSync('sample.txt', 'some more stuff\n', 'utf-8');

fs.appendFileSync('sample.txt', 'appended stuff', 'utf-8');
```

```typescript
// read from a file (Deno way)

for (const dir of Deno.readDirSync('.')) {
    if (dir.isDirectory) {
        console.log(dir.name);
    }
}

try {
    // generates an "exception" if not found
    Deno.statSync('emojys.ts');
} catch (error) {
    if (error instanceof Deno.errors.NotFound) {
        console.log('Couldn\'t find file.');
    }
}
try {
    Deno.removeSync('crappydb.bkp');
} catch {}

const records = Deno.readTextFileSync('crappydb.csv').split('\r\n');
for (const r of records.slice(0,records.length-1)) {
    const [id, first, last] = r.split(',');
    console.log(`${id}: ${last},${first}`);
}

Deno.writeTextFileSync('sample.txt', 'deno stuff for file');
Deno.writeTextFileSync('sample.txt', 'more deno stuff\n');

Deno.writeTextFileSync('sample.txt', 'deno appended stuff\n', {append:true});
```
