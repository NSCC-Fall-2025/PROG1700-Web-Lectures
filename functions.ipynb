```typescript
// a simple random number between 0 (inclusive) and 1 (exclusive)
Math.random()
```

```typescript
// a random number between 0 and 9 (inclusive)
Math.floor(Math.random() * 10 + 1)
```

```typescript
// a bunch of random numbers
for (let i = 0; i < 10; i++) {
    console.log(Math.floor(Math.random() * 10 + 1))
}
```

```typescript
let output = "";
for (let i = 0; i < 100; i++) {
    output += `${Math.floor(Math.random() * (20 - 10 + 1) + 10)} `;
}
console.log(output);
```

```typescript
// simple function

// declare the function
function writeHello() {
    console.log("Hello, World!");
    let sum = 1 + 2;
    console.log(sum);
}
// run the function
writeHello();
```

```typescript
// parameters
function writeHello(greeting: string, name: string) {
    console.log(`${greeting}, ${name}`);
}

writeHello("Hello", "World!");
writeHello("Bonjour", "Jaques!");
writeHello("Bonjourno", "Frederica!");
```
