```typescript
// return multiple values
function getMathAnswer(a: number, b: number) {
    const sum: number = a + b;
    const diff: number = a - b;
    const prod: number = a * b;
    const quot: number = a / b;

    return [sum, diff, prod, quot];
}

let [sum, difference, product, quotient] = getMathAnswer(1, 2);
console.log(`sum = ${sum}`);
console.log(`difference = ${difference}`);
console.log(`product = ${product}`);
console.log(`quotient = ${quotient}`);
```

```typescript
// change to use a single dictionary
function getMathAnswer({ a, b } : { number, number }) {
    const sum: number = a + b;
    const diff: number = a - b;
    const prod: number = a * b;
    const quot: number = a / b;

    return {sum: sum, difference: diff, product: prod, quotient: quot};
}

let answers = getMathAnswer({a: 1, b: 2});
console.log(`sum = ${answers.sum}`);
console.log(`difference = ${answers.difference}`);
console.log(`product = ${answers.product}`);
console.log(`quotient = ${answers.quotient}`);
```

```typescript
// pass by value
function passPrimitive(integer: number, decimal: number, bool: boolean, text: string) {
    integer++;
    decimal *= 37.0;
    bool = !bool;
    text = "a different string";
    console.log(`inside: i = ${integer}, d = ${decimal}, b = ${bool}, s = ${text}`);
}

let i: number = 10;
let d: number = 1.23;
let b: boolean = false;
let s: string = "a string";
console.log(`before: i = ${i}, d = ${d}, b = ${b}, s = ${s}`);
passPrimitive(i, d, b, s);
console.log(`after: i = ${i}, d = ${d}, b = ${b}, s = ${s}`);
```

```typescript
function passReference(arr: number[], dict: {[key:string]: string}) {
    arr[0] = 234;
    arr.push(999);
    arr = [6, 7, 8, 9, 10];
    dict.orange = 'orange';
    console.log(`inside: a = [${arr}], d = ${JSON.stringify(dict)}`);
}

let a = [1, 2, 3, 4, 5];
let d = {apple: 'red', banana: 'yellow'};
console.log(`before: a = [${a}], d = ${JSON.stringify(d)}`);
passReference(a, d);
console.log(`after: a = [${a}], d = ${JSON.stringify(d)}`);
```

```typescript
// default parameter values
function greeting(punctuation: string, target: string = "World!", hello: boolean = true) {
    if (hello) {
        console.log(`Hello, ${target}${punctuation}`);
    } else {
        console.log(`Goodbye, ${target}${punctuation}`);
    }
}

greeting("!");
greeting("?", "John");
greeting("!!!", "Jill", true);
//greeting(false); // can't skip over default parameters
```

```typescript
// optional parameters
function greeting(target: string, hello: boolean = "Hello", punctuation?: string) {
    const punct = punctuation || "!";
    if (hello) {
        console.log(`Hello, ${target}${punct}`);
    } else {
        console.log(`Goodbye, ${target}${punct}`);
    }
}

greeting("World");
```
